1. Read mobile dataset using pandas, and preprocess it (fill null values).

```
import numpy as np
import pandas as pd
a = pd.read_csv("/content/mobile_cleaned-1549119762886.csv")
print(a)
c_list = list(a.columns)
print(c_list)
for i in c_list:
  c = a[i].mean()
  a[i].isnull()
  a[i]=a[i].fillna(c)
  #print(c)

print(a)
```

2. Draw correlation graph for mobile price with all attributes.


```
import matplotlib.pyplot as plt
import pandas as pd
a = pd.read_csv("/content/mobile_cleaned-1549119762886.csv")
c1 = list(a.columns)
print(c1)
p = "+ve correlation "
n = "-Ve correlation "
no = "No correlation "
c2_list = []
for q in range(39):
  if q == 2 or q == 27 or q == 37 :
    c2_list.append(n)
  elif q == 30 or q == 34 :
    c2_list.append(p)
  else:
    c2_list.append(no)
print(c2_list)
c4 = 0
for i in c1:
  if i != "price":
    ax1 = a.plot.scatter(x="price",y=i,c='DarkBlue')
    plt.title("correlation between price and "+i)
    plt.text(5,5,c2_list[c4]+"between "+i+" and price")
    c4 +=1
```

3. Compute outer product of 2 vectors.

```
import numpy as np
p =[[1,0],[0,1]]
q =[[1,2],[3,4]]
r = np.outer(p,q)
print("Original matrix")
print(p)
print(q)
print("ouer product of the said two vectors")
print(r)
```

4.  Write a NumPy program to compute the condition number of a given matrix.

```
import numpy as np
from numpy import linalg as la
a = np.array([[1,2],[3,4]])
print("Original matrix:")
print(a)
print("The condition number of the said matrix:")
print(la.cond(a))
```
